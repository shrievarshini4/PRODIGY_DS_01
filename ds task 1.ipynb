```python
import pandas as pd

# Load the dataset
data_url = r"C:\Users\shrie\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_45183\API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv"
df = pd.read_csv(data_url, skiprows=3)

# Display the first few rows of the dataset
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt



# Step 2: Filter data for six specific countries
countries_of_interest = ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 'Brazil']
filtered_df = df[df['Country Name'].isin(countries_of_interest)]

# Step 3: Extract country names and total population for the selected countries
countries = filtered_df['Country Name']
total_population = filtered_df['2021']  # Assuming you want to plot the data for the year 2019

# Step 4: Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(countries, total_population, color='skyblue')
plt.title('Total Population by Country (2021)')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
